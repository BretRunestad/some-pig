```python
class Player:
    """
    Responsibilities:
    * keeps track of points
    * has a strategy
    """
    def __init__(self):
        pass
        
    def will_roll(self, roll, roll_points):
        if roll == 1:
            return True
        else:
            return False
        
```

```python
class Gordon(Player):
    def will_roll(self, roll, roll_points):
        return True
```

```python
class Random(Player):
    def will_roll(self, roll, roll_points):
        if random.choice([0,1]) == 1:
            return True
        else:
            return False
```

```python
class Game:
    """
    Responsibilities:
    * rolls dice
    * handles outcome of dice rolls
    * iterates over player rolls in turns
    * iterates over turns for game
    * 
    
    Collaborators:
    * Player
    """
    def __init__(self, player):
        self.turn = 0
        self.player = player
        self.player_score = 0
    
    def dice_roll(self):
        return random.randint(1, 6)
    
    def roll_success(self, die):
        if die == 1:
            return False
        else:
            return True
    
    def turn_event(self, player):
        #turn_points = 0
        roll = 1
        roll_points = 0
        while True:
            if player.will_roll(roll, roll_points):
                die = self.dice_roll()
                if self.roll_success(die):
                    roll_points += die
                else:
                    return 0
                roll += 1    
            else:
                return roll_points
        
                
            #roll += 1
        #turn_points += roll_points
        #print(turn_points)
        #return turn_points
        
    def complete_game(self):
        for _ in range(7):
            self.player_score += self.turn_event(player)
        return self.player_score
```

```python
player = Player()
game = Game(player)
print(game.complete_game())
```

```python
def multiple_games(n, player):
    score_list = []
    for _ in range(n):
        game = Game(player)
        score_list.append(game.complete_game())
    return(score_list)

player = Player()
print(multiple_games(10, player))
```

```python
basic_player = Player()
basic_player_1000 = multiple_games(1000, player)
basic_player_mean = st.mean(basic_player_1000)
basic_player_stdev = st.stdev(basic_player_1000)
print(basic_player_mean)
print(basic_player_stdev)
```

```python
plt.hist(basic_player_1000, bins= 10)
plt.axvline(x=basic_player_mean, color = 'red')
plt.axvline(x=basic_player_mean+basic_player_stdev, color = 'red', linestyle='dashed')
plt.axvline(x=basic_player_mean-basic_player_stdev, color = 'red', linestyle='dashed')
plt.figtext(x=.67, y=.75, s="mean  =  23.431")
plt.figtext(x=.67, y=.65, s="std dev =  5.18576")
plt.xlabel("points")
plt.ylabel("frequency")
plt.title("Scores of Basic Player")
plt.show()
```
